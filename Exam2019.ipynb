```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
%matplotlib inline
```

```python
X = np.arange(-10, 10, 0.2)
Y = np.arange(-10, 10, 0.2)
X, Y = np.meshgrid(X, Y)
X_len,Y_len = len(X)//2 ,len(Y)//2
```

```python
def rotate(image, angle, center=None, scale=1.0):
    # angle 负是顺时针旋转
    (h, w) = image.shape[:2]
    if center is None:
        center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, scale)
    rotated = cv2.warpAffine(image, M, (w, h))
    return rotated
```

```python
Z = np.exp(-5/16*X**2 +3*np.sqrt(3)/8*X*Y -15/16*Y**2)
plt.figure(figsize=(15,10))

plt.figure()
plt.subplot(121)
plt.contour(X,Y,Z)

plt.subplot(122)
angle = np.arctan(3*np.sqrt(3)/5)*(180/np.pi)/2

plt.contour(X,Y,rotate(Z, angle, center=None, scale=1.5))
plt.show()
```
